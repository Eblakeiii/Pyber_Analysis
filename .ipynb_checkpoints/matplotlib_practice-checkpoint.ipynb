```python
%matplotlib inline

# dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.xlabel("Months")
plt.ylabel("Amount")
plt.title("First Graph Attempt")
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# skills drill - Create the plot.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("Fares by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "Arp", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09 ]
```

```python
# Create the plot for bar chart
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot for horizontal bar chart
plt.barh(x_axis, y_axis, color="pink", label='Boston')
# labels for axis
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("Fares by Month in Boston")
plt.legend()
```

```python
# Create the plot - invested axis
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()
# labels for axis
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("Pyber Fares by Month in Boston")
plt.legend()
```

```python
# skills drill Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="black")
ax.set_ylabel("Fare($)")
ax.set_xlabel("Date")
ax.set_title("Pyber Fares in New York")
ax.legend("New York")
```

```python
# skills drill Create the object oriented plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis, color="cyan")
ax.set_xlabel("Date")
ax.set_ylabel("Amount")
ax.set_title("Pyber Fares by Month in Chicago")
ax.legend(["Chicago"])
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
# skills drill
plt.scatter(y_axis, x_axis, color="red", label='Chicago')
plt.gca().invert_yaxis()
plt.xlim(0,45)
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("Pyber Fares by Month in Chicago")
plt.legend()
```

```python
# bubble chart
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

```python
# object oriented-approach to scatter plots
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], color="skyblue", alpha=0.2, edgecolor="k", linewidth=2)
plt.gca().invert_yaxis()
ax.set_xlim(0, 45)
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.set_title("Fares in Boston")
ax.legend(["Boston"])

# , linewidth=1
#linestyle='dashed'
```

```python
# create pie chart
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Assign 12 colors, one for each month; explode a wedge; label with percents showing
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# skills drill - create pie chart using object-oriented interface
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%', shadow=True, startangle=90, counterclock=False)

plt.show()
```
